```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"id" : [1,2,3,4],"name" : ['A','B','C','D']})
df1
```

```python
df11 = pd.DataFrame({"id" : [1,2,3,4],"name" : ['A','B','C','D']})
df11
```

```python
df2 = pd.DataFrame({"id" : [1,2,3,4],"class" : ['six','seven','eight','nine']})
df2
```

```python
df3 = pd.DataFrame({"id" : [1,2,3,6],"home" : ['chapai','raj','mohonpur','diar']})
df3
```

```python
df4 = pd.DataFrame({"id" : [5,6,7,8],"mobile" : ['017','019','016','015']})
df4
```

```python
merge1 = pd.merge(df1, df2, on="id")
merge1
```

```python
merge1 = pd.merge(df2, df1, on="id")
merge1
```

```python
merge3 = pd.merge(df3, df1, on="id")
merge3
```

```python
merge_inner = pd.merge(df1,df3, on='id', how='inner')
merge_inner
```

```python
merge_left = pd.merge(df1,df3, on='id', how='left')
merge_left
```

```python
merge_right = pd.merge(df1,df3, on='id', how='right')
merge_right
```

```python
merge_outer = pd.merge(df1,df3, on='id', how='outer')
merge_outer
```

```python
merge_outer_indicator = pd.merge(df1,df3, on='id', how='outer', indicator=True)
merge_outer_indicator
```

```python
merge_different_id = pd.merge(df1, df4, left_index=True, right_index=True)
merge_different_id
```

```python
merge_tuple = pd.merge(df1, df11, on="id")
merge_tuple
```

```python
merge_tuple = pd.merge(df1, df11, on="id", suffixes=(['_higher','_moddle']))
merge_tuple
```
