```python
import pandas as pd
```

```python
sr1 = pd.Series([1,2,3]) ; sr2 = pd.Series([4,5,6])
print(sr1) 
print(sr2)
```

```python
concat1 = pd.concat([sr1,sr2])
concat1
```

```python
concat1 = pd.concat([sr2,sr1])
concat1
```

```python
concat1 = pd.concat([sr2,sr1],ignore_index=True)
concat1
```

```python
df1 = pd.DataFrame({'id': [1,2,3], 'name' : ['abir','jabir','khabir'], 'class': [6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'id': [4,5,6], 'name' : ['saddam','monir','titumir'], 'class': [10,11,12]})
df2
```

```python
df3 = pd.DataFrame({'id':[110,111], 'name': ['sabbir','masud'], 'class' : [14,15]})
df3
```

```python
concat1 = pd.concat([df1,df2])
concat1
```

```python
concat1 = pd.concat([df2,df1])
concat1
```

```python
concat1 = pd.concat([df2,df1], ignore_index = True)
concat1
```

```python
concat1 = pd.concat([df1,df2], axis = 0)
concat1
```

```python
concat1 = pd.concat([df1,df2], axis =1)
concat1
```

```python
concat1 = pd.concat([df1,df3],axis =0)
concat1
```

```python
concat1 = pd.concat([df1,df3],axis =1)
concat1
```

```python
concat1 = pd.concat([df1,df3],axis =1, join = 'inner')
concat1
```

```python
concat1 = pd.concat([df1,df3],axis =1, join= 'outer')
concat1
```

```python
concat1 = pd.concat([df1,df3], keys =['df1', 'df3'] )
concat1
```

```python
concat1 = pd.concat([df1,df3], keys =['first df', 'second df'] )
concat1
```

```python
concat1 = pd.concat([df1,df3],axis =1, keys =['first df', 'second df'] )
concat1
```
