```python
import pandas as pd
```

```python
df1 = pd.read_excel('for pivot tabel.xlsx')
df1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre')
pivot_tabel1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre', columns='Year')
pivot_tabel1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre', columns='Year', aggfunc='count')
pivot_tabel1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre', columns='Year', aggfunc='max')
pivot_tabel1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre', columns='Year', aggfunc='sum')
pivot_tabel1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre', columns='Year', aggfunc='sum', fill_value=0)
pivot_tabel1
```

```python
pivot_tabel1 = df1.pivot_table(index='Genre', columns='Year', aggfunc='sum', fill_value=0, margins=True)
pivot_tabel1
```
